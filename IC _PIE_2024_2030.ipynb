```python
# Inserting Columns for Total Thermal and Total Renewables Color_Source= {'Thermal':'lightgray','Coal & Lignite':'lightgrey', 'Gas': 'gray',  
               'Nuclear': 'pink', 'Hydro': 'aquamarine', 'HYDRO': 'aquamarine',
               'Renewables': 'springgreen', 'PSP':'lime',
               'Wind': 'limegreen' ,'Small Hydro': 'royalblue', 'Biomass':'silver','Solar': 'yellow'}
df_AI['Thermal']= df_AI[TH].sum(axis=1)
df_AI['HYDRO']= df_AI[HY].sum(axis=1) # HYDRO includes Hydro and PSP
df_AI['Renewables']= df_AI[RE].sum(axis=1)
```

```python
Sources = ['Thermal', 'Nuclear', 'HYDRO', 'Renewables' ]
df_AI['Total IC']= df_AI[Sources].sum(axis=1)
```
